```python
class DiffieHellman:
    def __init__(self, prime, primitive_root):
        self.prime = prime
        self.primitive_root = primitive_root

    def generate_public_key(self, private_key):
        return pow(self.primitive_root, private_key, self.prime)

    def calculate_shared_secret(self, public_key, private_key):
        return pow(public_key, private_key, self.prime)

#paramètres
prime = 47
primitive_root = 2
dh = DiffieHellman(prime, primitive_root)

#clés Privées
private_key_lincoln = 8
private_key_washington = 15

#calcul des clés publiques
public_key_lincoln = dh.generate_public_key(private_key_lincoln)
public_key_washington = dh.generate_public_key(private_key_washington)

#calcul des clés secrètes partagées
shared_secret_lincoln = dh.calculate_shared_secret(public_key_washington, private_key_lincoln)
shared_secret_washington = dh.calculate_shared_secret(public_key_lincoln, private_key_washington)

print("Shared Secret (Lincoln's side):", shared_secret_lincoln)
print("Shared Secret (Washington's side):", shared_secret_washington)
```
